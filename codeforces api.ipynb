```python
def make_problems_list(list, data): # Lọc key Problems và thêm key Verdict vào key Problems
    for bit in data:
        bit['problem']['verdict'] = bit['verdict']
        bit['problem']['creationTimeSeconds'] = bit['creationTimeSeconds']
        list.append(bit['problem'])
```

```python
user_problems = []

make_problems_list(user_problems, user_result)

user_problems
```

```python
def make_problems_sovled_list(sovled_list, data): # Lọc những problem đã AC (kể cả các probelms AC nhiều lần)
    user_problems = []
    make_problems_list(user_problems, data)
    for problem in user_problems:
        if(problem['verdict'] == 'OK'):
            sovled_list.append(problem)
```

```python
user_problems_sovled = []

make_problems_sovled_list(user_problems_sovled, user_result)

print( len(user_problems_sovled),'\n')
user_problems_sovled
```

```python
# Loại bỏ những submit trùng problem
user_problems_sovled = list({problem['name']: problem for problem in user_problems_sovled}.values())

user_problems_sovled
```

```python
problems_list = {}
for i in user_problems_sovled:
    problems_list[i['name']] = i['index']
#problems_list = set(problems_list)
problems_list
```

```python
indexes = problems_list.values()
cnt = {}
for index in indexes:
    if index not in cnt.keys():
        cnt[index] = 1
    else: 
        cnt[index] += 1
cnt
    
```

```python
values = list(cnt.values())
names = list(cnt.keys())
plt.figure(dpi = 80)
plt.bar(range(len(cnt)), values, tick_label = names)

plt.title('Problems sovled by index')
plt.show()
```

```python
plt.figure(dpi = 80)
plt.pie(values, labels = names, radius = 1.8, counterclock = False, autopct='%1.0f%%', pctdistance=1.1, labeldistance=1.2)
plt.show()
```

```python
time_in_unix = []
for problem in user_problems_sovled:
        time_in_unix.append(problem['creationTimeSeconds'])
time_in_unix
```

```python
from datetime import datetime
```

```python
def unix_to_datetime(unix):
    unix = datetime.fromtimestamp(unix)
    return unix
```

```python
date_time = []
for time in time_in_unix:
    time = unix_to_datetime(time)
    date_time.append(time)
date_time
```

```python
monthly_sovled = {}
for i in range(1,13):
    monthly_sovled[i] = 0
for time in date_time:
        monthly_sovled[time.month] +=1
monthly_sovled
```

```python
monthly_sovled = dict(sorted(monthly_sovled.items()))
monthly_sovled
```

```python
x = list(monthly_sovled.keys())
y = list(monthly_sovled.values())
plt.plot(x, y, marker = 'o')
plt.xticks(x)
#plt.figure(dpi = 80)
plt.xlim(0,12)
plt.title('Problems sovled monthly')
plt.show()
```

```python
problems_tags = {}
for problem in user_problems_sovled:
    tags = problem['tags']
    for tag in tags:
        if tag not in problems_tags:
            problems_tags[tag] = 0
        else:
            problems_tags[tag] +=1
problems_tags = dict(sorted(problems_tags.items(), key = lambda item : item[1], reverse = True))
problems_tags
```

```python
x_tag = list(problems_tags.keys())
y_tag = list(problems_tags.values())
plt.figure(dpi = 100)
plt.pie(y_tag, counterclock = False, radius = 1.5, startangle = 90)
plt.legend(x_tag, loc = 'best', bbox_to_anchor=(2, 1))
#plt.axis('equal')
#plt.tight_layout()
plt.show()
```
